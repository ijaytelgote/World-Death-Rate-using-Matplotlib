```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Disease=['Corona','Dengue','Typhoid','Cancers','Sugar','Other']
Percentage=[46,13,6.4,16.8,7,10.8]
```

```python
fig=plt.figure(figsize=(8,8))    #setting the pie chart #set figure size
    #exploring the largest pie in data chart
xplode = (0, 0, 0, 0, 0.2)
plt.pie(Percentage, labels = Disease,autopct='%1.1f%%',explode=None,startangle=90)  
#setting axis equal to pie as circle

plt.title('World Death Rate 2020')
plt.show()  
```
